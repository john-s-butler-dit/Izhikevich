## The Izihkevich Model
The Izihkevich model is a system of two first order non-linear differential equations:
\begin{equation}  \frac{dv}{dt} = 0.04v^2+5v+140-u+I, \end{equation}
\begin{equation}\frac{du}{dt} = a(bv-u),\end{equation}.

Where $u$ and $v$ are dimensionless variables, $I$ is the input to the system and $a$ and $b$ are dimesionless parameters which are maniputled to descibe different firing patterns (see next section). 

The variable $v$ is of most interest as it represents the membrane potential of the neuron in $mV$ and $u$ represents a
membrane recovery variable, which accounts for the activation of $K^{+}$
ionic currents and inactivation of $Na^{+}$ ionic currents, and it provides
negative feedback to $v$. 

After the spike reaches its apex (30 mV), the
membrane voltage and the recovery variable are reset according  
if $v\geq 30mV$ then
\begin{equation}  v \leftarrow c, \end{equation}
\begin{equation} u \leftarrow u+d.\end{equation}

### Parameters
By changing parameters $a$, $b$, $c$ and $d$ different firing patterns can be simulated. Each parameter corresponds to different aspects of neurnal behaviour:

a. The parameter  $a$ describes the time scale of the recovery variable $u$.
Smaller values result in slower recovery. 

b. The parameter $b$ describes the sensitivity of the recovery variable
to the subthreshold fluctuations of the membrane potential $v$.
Greater values couple and more strongly resulting in possible
subthreshold oscillations and low-threshold spiking dynamics. 

c. The parameter $c$ describes the after-spike reset value of the membrane
potential $v$ caused by the fast high-threshold $K^{+}$ conductances.

d. The parameter $d$ describes after-spike reset of the recovery variable $u$
caused by slowhigh-threshold $Na^{+}$ and $K^{+}$conductances.

## Methods
### Numerical Solution
The system of differential equations describe in the previous section is non-linear and has no known exact solution, the solution has to be numerically approximated. The simplest approximation method is the Euler method (https://en.wikipedia.org/wiki/Euler_method) which approximates the first derivate by the formula of the slope of a line at two time points $t$ and $t+h$, where $h$ is a small step in time:
\begin{equation}  \frac{dv}{dt} \approx \frac{v(t+h)-v(t)}{t+h-t}=\frac{v(t+h)-v(t)}{h}. \end{equation}
Using this method the Izhikecich equations can be re-written as 
\begin{equation}   \frac{v(t+h)-v(t)}{h} \approx 0.04v^2+5v+140-u+I, \end{equation}
\begin{equation} \frac{u(t+h)-u(t)}{h} \approx a(bv-u)
.\end{equation}
By re-arranging these equations to  numerically approximate the solution $v(t+h)$ and $u(t+h)$ at time $t+h$ using the solutions $u(t)$ and $v(t)$  and input $I(t)$ at time $t$, such that,
\begin{equation}   v(t+h) \approx v(t)+h(0.04v(t)^2+5v(t)+140-u(t)+I(t)), \end{equation}
\begin{equation} u(t+h)\approx u(t)+h(a(bv(t)-u(t))).\end{equation}
The smaller the time step $h$ between two time points the more accurate the method, here $h=0.5ms$.

In the cell below is the function DiscreteModel is the numerical Differential Equation which gives the approximate numerical solution of $u$, and $v$ at time point $t+h$. The function takes in the parameters $a$ and $b$, and the approximate numerical solution of $u$, and $v$ at time point $t$. 

```python
# Numerical Solution to the Izhikevich model
def DiscreteModel(a,b,u,v,I):
    v = v + 0.5*(0.04*v*v+5*v+140-u+I) # Discrete form of membrane potential of the neuron 
    v = v + 0.5*(0.04*v*v+5*v+140-u+I) # Discrete form of membrane potential of the neuron 
    u = u + (a*(b*v-u))            # Discrete form of membrane recovery variable
    return u,v      
```

### The main code
In the cell below is the main function for the Izhikevich model. The function takes in four variables a, b ,c and d coresponding to the parameters in to model of the same namesake. 

```python
#def Izhikevich(a,b,c,d):
n=1000 # Number of Neurons
nI=200
nE=n-nI
```

```python
re=np.random.rand(nE)
ri=np.random.rand(nI)
a=np.concatenate([0.02*np.ones(nE),0.02*np.ones(nI)+0.08*ri])  
b=np.concatenate([0.2*np.ones(nE),0.25*np.ones(nI)-0.05*ri])  
c=np.concatenate([-65*np.ones(nE)+15*re**2,-65*np.ones(nI)])  
d=np.concatenate([8*np.ones(nE)-6*re**2,2*np.ones(nI)])
time=np.arange(0,2000.1,1)
lfp=0*np.ones((len(time)))
```

## Neuronal Connectivity
Each neuron (i) is connected to another neurons (j) by a connection weight ($S_{ij}$) that is a positive value in the range of $[0, 0.5]$ for an exhitatory connection and negative value in the range $[0,-1]$ for an inhibitory connection. These weights are defined in a 1000 $\times$ 1000 matrix $S$.  The cell bellow random assigns connection weights to each neuron and plots the corresponding connectivity matrix.

```python
S=np.hstack((0.5*np.random.rand(n,nE),-np.random.rand(n,nI)))
#S=np.hstack((0.5*np.ones((n,nE)),-np.ones((n,nI))))


fig, ax = plt.subplots(figsize=(10,7))

im =plt.imshow(S.transpose())#, cmap=cm.RdBu, vmin=abs(S).min(), vmax=abs(S).max())
plt.xlabel('Neuron')
plt.ylabel('Neuron')
plt.title('Connectivity Matrix')
im.set_interpolation('bilinear')

cb = fig.colorbar(im)
```

```python
v=-65*np.ones((len(time),n))  # SETS UP THE INITIAL VALUES of the membrane potential
u=0*np.ones((len(time),n))    # SETS UP THE INITIAL VALUES of the membrane recovery
U=b*c                 # INITIAL CONDITION 
fired=0*np.ones(1)
FIRED=0*np.ones(n);
V=v[0,:]
INTERNAL= np.dot(S,FIRED)
```

```python
firing=[]
firing_time=[]
index_list=np.arange(n)

for k in range (0,len(time)-1):
    INPUT= np.concatenate([5*np.random.randn(nE),2*np.random.randn(nI)])+INTERNAL.transpose()
    U,V=DiscreteModel(a,b,U,V,INPUT) # Numerically solving the equations at each time point
    FIRED[V>30]=1 
    INTERNAL= np.dot(S,FIRED)
    if sum(FIRED)>0:
        firing.extend(index_list[FIRED==1])
        firing_time.extend(time[k]*np.ones(len(c[FIRED==1])))
        V[FIRED==1]=c[FIRED==1]
        U[FIRED==1]=U[FIRED==1]+d[FIRED==1]
        lfp[k]=FIRED.sum()
        FIRED=0*np.ones(n)
        
   
    u[k+1,:]=U
    v[k+1,:]=V
    
```

```python
## PLOTTING
fig, ax1 = plt.subplots(figsize=(10,7))

ax1.plot(firing_time, firing, '.')
ax1.set_xlabel('time (ms)')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('Neuron', color='b')
ax1.tick_params('y', colors='b')
fig.tight_layout()
ax1.legend(loc=1)


fig, ax1 = plt.subplots(figsize=(10,3))
plt.plot(time,lfp)
plt.xlabel('time (ms)')
plt.ylabel('Number of Spikes ')
plt.show()
```
